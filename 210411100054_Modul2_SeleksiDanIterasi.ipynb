####**1. Terdapat tiga syntax percabangan di dalam Python, yaitu if, if else, dan if elif .. else. Jelaskan perbedaan ketiganya, dan tuliskan contoh code masing-masing syntax**


**Perbedaan if, if else, if elif .. else**
*   **if :** percabangan tunggal, statement akan dieksekusi jika bernilai True,
sebaliknya jika bernilai False maka statement tidak akan dieksekusi
*   **if else :** jika stetement bernilai True maka akan dieksekusi dalam kode **if**, namun jika statement bernilai false maka akan dieksekusi di kode **else**
*   **if elif .. else :** kondisi ini merupakan lanjutan dari **if**. Kondisi **elif** membuat program dapat menyeleksi beberapa kemungkinan yang dapat terjadi. hampir sama dengan **else**, bedanya kondisi **elif** bisa lebih dari satu





**Contoh syntax if**

```python
# contoh 1
angka = int(input("masukkan angka = "))
if angka % 2 == 0:
  print(angka, "adalah bilangan genap")
```

```python
# contoh 2
angka = int(input("masukkan angka = "))
if angka % 2 == 0:
  print(angka, "adalah bilangan genap")
```

**Contoh syntax if else**

```python
# contoh 1
angka = int(input("masukkan angka = "))
if angka % 2 == 0:
    print(angka, "adalah bilangan genap")
else:
    print(angka,"adalah angka ganjil")
```

#### **2. Terdapat dua syntax perulangan di dalam Python, yaitu for dan while. Jelaskan perbedaan keduanya,dan tuliskan contoh code masing-masing syntax**

**perbedaan for dan while**


*   **for :** perulangan akan terus berjalan selama kondisi bernilai True dan perulangan akan berhanti jika kondisi bernilai False atau sudah tidak terpenuhi lagi. Perulangan **for** dapat mengulang item dari urutan apapun, contohnya list atau string
*   **while :**pengulangan **while** akan mengeksekusi stetement berkali-kali selama kondisinya bernilai True



**Contoh syntax for**

```python
# contoh 1
angka_range = range(10)

for i in angka_range:
    print("angka sekarang menjadi = ",i)
print("akhir dari program")
```

```python
# contoh 2
data_str = "saya adalah aini"

for huruf in data_str:
    print(huruf)
print("akhir dari program")
```

**Contoh syntax while**

```python
angka = 0
print("angka sekarang",angka)

while angka < 5:
    angka += 1
    print("angka sekarang menjadi",angka)

print('akhir dari program')
```

berupa jumlah bab dan sub bab dalam setiap bab. [Hint : Buatlah iterasi di dalam iterasi]

```python
bab = int(input('Masukkan jumlah bab = '))
subbab = int(input('Masukkan jumlah sub bab di setiap bab = '))

for x in range(1, bab+1):
    print(x,'.')
    for y in range(1, subbab+1):
        print(' ', x, '.', y)
```

#### **2.3 Menu**



---


Buatlah code untuk membuat menu, dimana setiap menu melakukan operasi yang berbeda :


Menu :

*   Tekan 1 untuk operasi perhitungan luas lingkaran (input adalah jari-jari)
*   Tekan 2 untuk operasi perhitungan luas persegi panjang (input adalah panjang dan lebar)
*   Tekan 3 untuk operasi perhitungan luas segitiga (input adalah alas dan tinggi)

Tambahkan pilihan bagi user, apakah ingin mengulangi operasi kembali ataukah tidak, jika ‘y’, maka
tampilkan menu kembali dan lakukan operasi sesuai pilihan dari user. Jika user memilih ‘t’ maka operasi
berhenti. Contoh output dari code tersebut dapat dilihat pada Gambar 3


---
Screenshot 2021-11-18 105544.png




```python
nilai = True
while nilai :
    print("Menu")
    print("Tekan 1 untuk operasi perhitungan luas lingkaran")
    print("Tekan 2 untuk operasi perhitungan luas persegi panjang")
    print("Tekan 3 untuk operasi perhitungan luas segitiga")
    cek = int(input("massukkan pilihan anda = "))
    if cek == 1:
        r = int(input("masukkan jari-jari = "))
        luas = 22/7 *r*r
        print("Luas lingkaran = ", luas)
    elif cek == 2 :
        p = int(input("masukkan panjang = "))
        l = int(input("masukkan lebar = "))
        luas = p * l
        print("Luas persegi panjang = ", luas)
    elif cek == 3 :
        a = int(input("masukkan alas = "))
        t = int(input("masukkan tinggi = "))
        luas = 1/2 * a * t
        print("Luas segitiga = ", luas)
    cek = str(input('Ingin mengulang operasi kembali (y/t) ? '))
    print(20*'=')
    if cek == 'y' :
        nilai = True
    elif cek == 't' :
        nilai = False
```
