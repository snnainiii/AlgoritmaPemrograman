#### **1. Pengertian List, dan berikan contoh**

#### **2. Pengaksesan variabel yang bertipe List (jelaskan indeks, dan pengaksesan setiap anggota yang terdapat pada list), dan berikan contoh**


*   **pengertian index**

    index adalah cara mendapatkan atau mengetahui informasi mengenai posisi elemen di dalam list

*   **pengaksesan setiap anggota yang terdapat pada list**

  [n] = menampilkan suatu index

  [a:b] = menampilkan index dari a sampai b-1

  [a:] = menampilkan index dari sesudah a sampai akhir

  [:b] = menampilkan index dari awal sampai b

  [-n] = menampilkan index dari element terakhir/index dimulai dari element terakhir

  [:-n] = menampilkan index dari a sampai sebelum terakhir

  [:] = menampilkan semua index


*   **contoh**




```python
# contoh [n]
nilai = [60, 65, 70, 75, 80, 85, 90, 95, 100]
print(nilai[3])
print(nilai[5])
print(nilai[8])
```

```python
# contoh [a:b]
nilai = [60, 65, 70, 75, 80, 85, 90, 95, 100]
print(nilai[4:8])
print(nilai[2:9])
print(nilai[1:6])
```

```python
# contoh [a:]
nilai = [60, 65, 70, 75, 80, 85, 90, 95, 100]
print(nilai[3:])
print(nilai[8:])
print(nilai[5:])
```

```python
# contoh [:b]
nilai = [60, 65, 70, 75, 80, 85, 90, 95, 100]
print(nilai[:4])
print(nilai[:1])
print(nilai[:8])
```

```python
# contoh [-n]
nilai = [60, 65, 70, 75, 80, 85, 90, 95, 100]
print(nilai[-1])
print(nilai[-2])
print(nilai[-5])
```

```python
# contoh [:-n]
nilai = [60, 65, 70, 75, 80, 85, 90, 95, 100]
print(nilai[:-1])
print(nilai[:-3])
print(nilai[:-6])
```

#### **2.1 Faktor Pembagi**
Buatlah code dengan menggunakan Python untuk mendapatkan faktor pembagi dari suatu bilangan, dengan ketentuan sebagai berikut:


*   Bilangan yang akan dicari faktor pembaginya merupakan inputan dari user
*   Masukkan faktor pembagi yang didapat di dalam suatu list


Contoh output yang dihasilkan dapat dilihat pada Gambar 1.
Screenshot 2021-12-01 180251.png

```python
# gambar 1a
bil = int(input('masukkan bilangan = '))
hasil = []
pembagi = 0
for i in range(1, bil + 1):
    if bil % i == 0 :
        hasil.append(i)
        pembagi += 1
print('Bilangan', bil, 'memiliki', pembagi, 'faktor pembagi, yaitu')
print('faktor pembagi : ', hasil)
```

#### **2.2 Faktor Persekutuan Terbesar / FPB**
Buatlah code dengan menggunakan Python untuk mendapatkan Faktor Persekutuan Terbesari dari dua buah bilangan, dengan ketentuan sebagai berikut:


*   Dua Bilangan yang akan dicari FPB-nya merupakan inputan dari user
*   Simpan Faktor pembagi dari masing-masing bilangan ke dalam list
*   Simpan Faktor Pembagi yang sama dari dua bilangan tersebut
*   Dapatkan FPB dari dua bilangan tersebut

Contoh output yang dihasilkan dapat dilihat pada Gambar 2
Screenshot 2021-12-01 180727.png

```python
# gambar 2a
bil1 = int(input('Masukkan bilangan pertama = '))
bil2 = int(input('Masukkan bilangan pertama = '))
hasil1 = []
hasil2 = []
hasilakhir = []
for bagi1 in range(1, bil1 + 1):
    if bil1 % bagi1 == 0:
        hasil1.append(bagi1)
print('faktor pembagi', bil1, '=', hasil1)

for bagi2 in range(1, bil2 + 1):
    if bil2 % bagi2 == 0:
        hasil2.append(bagi2)
print('faktor pembagi', bil2, '=', hasil2)

if bil1 > bil2 :
    x = bil2
else :
    x = bil1

for sama in range(1, x + 1):
    if bil1 % sama == 0 and bil2 % sama == 0 :
        hasilakhir.append(sama)
        fpb = sama
print('Pembagi yang sama = ', hasilakhir, '; FPB = ', fpb)
```

#### **2.3 Nilai Mahasiswa**
Buatlah code untuk membuat daftar nilai mahasiswa dengan ketentuan sebagai berikut :


*   Buat inputan untuk menginputkan data mahasiswa berupa nama dan nilai, sesuai dengan jumlah
mahasiswa yang diinputkan, seperti Gambar 3a
*   Buat Menu Daftar Nilai Mahasiswa yang memiliki operasi-operasi sebagai berikut :

    – Daftar keseluruhan data mahasiswa (nama dan nilai)

    – Perhitungan nilai rata-rata mahasiswa
    – Daftar mahasiswa yang memiliki nilai lebih dari threshold (dimana nilai threshold merupakan inputan dari user)

    – Nilai maksimal

Contoh output dapat dilihat pada Gambar 3.
Screenshot 2021-12-01 180805.png

Screenshot 2021-12-01 180838.png

Screenshot 2021-12-01 180918.png

```python
# gambar 3a
nama = []
nilai = []
mhs = int(input('masukkan jumlah mahasiswa = '))
for i in range(1, mhs+1):
    print('Mahasiswa ke-', i)
    x = input('Masukkan Nama Mahasiswa = ')
    nama.append(x)
    y = int(input('Masukkan Nilai Mahasiswa = '))
    nilai.append(y)
```

```python
# gambar 3b
nama = []
nilai = []
mhs = int(input('masukkan jumlah mahasiswa = '))
for i in range(1, mhs+1):
    print('Mahasiswa ke-', i)
    x = input('Masukkan Nama Mahasiswa = ')
    nama.append(x)
    y = int(input('Masukkan Nilai Mahasiswa = '))
    nilai.append(y)
mulai = True
while mulai:
    print('Daftar Nilai Mahasiswa')
    print('Tekan 0 untuk Daftar keseluruhan mahasiswa dan nilainya')
    print('Tekan 1 untuk perhitungan rata rata')
    print('Tekan 2 untuk daftar mahasiswa dengan nilai dari thereshold')
    print('Tekan 3 nilai tertinggi')
    pilih = int(input('masukkan pilihan anda = '))
    if pilih == 0:
        for i in range(0, mhs):
            print(i+1, '.', nama[i], ':', nilai[i])
    elif pilih == 1:
        total = 0
        for i in nilai :
            total += i
        rata = total/mhs
        print('rata-rata nilai = ',rata)
    elif pilih == 2:
        a = 0
        nilaimin = int(input('masukkan nilai minimal = '))
        for i in range(mhs):
            if nilai[i] > nilaimin :
                a += 1
                print(a, '.', nama[i],'; nilai = ', nilai[i])
    elif pilih == 3:
        max = nilai[0]
        for i in range(mhs):
            if nilai[i] > max:
                max = nilai[i]
        print('nilai Tertinggi = ', max)
    ulang = input('Ingin mengulang operasi kembali (y/t) ? ')
    if ulang == 'y':
        mulai = True
        print(70*'=')
    else:
        mulai = False
        print(70*'=')
```
