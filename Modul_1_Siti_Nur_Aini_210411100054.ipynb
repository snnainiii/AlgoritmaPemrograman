## 2.1 Algoritma Bilangan Ganjil


---


Tulis tahapan-tahapan (algoritma) untuk menampilkan sejumlah n bilangan ganjil, dimana n adalah inputan
dari user, dan jumlahkan semua bilangan ganjil tersebut.


---

1. meminta pengguna memasukkan bilangan bulat (bilanganmax)
2. buat variabel b = 0
3. buat variabel hasil = 0
4. lakukan perulangan dengan range(bilanganmax*2)
5. buat pengkondisian i % 2 == 1
6. lakukan penugasan b += 1
7. lakukan penugasan hasil += i
8. kemudian print("bilangan ke- ", b , ":", i)
9. tampilkan hasil

## 2.2 Program Bilangan Ganjil


---


Konversi algoritma yang sudah ditulis sebelumnya dengan menggunakan bahasa Pemrograman Python.
Contoh hasil eksekusi dapat dilihat di Fig.1a or Fig.1b :


---
Screenshot (200).png


```python
bilanganmax = int(input("masukkan jumlah bilangan = "))
b = 0
hasil = 0
for i in range(bilanganmax*2):
  if i % 2 == 1 :
    b += 1
    hasil += i
    print("bilangan ke-", b , ":", i)
print("total =", hasil)
```

```python
bilanganmax = int(input("masukkan jumlah bilangan = "))
b = 0
hasil = 0
for i in range(bilanganmax*2):
  if i % 2 == 1 :
    b += 1
    hasil += i
    print("bilangan ke-", b , ":", i)
print("total =", hasil)
```

## 2.4 Program Deret Aritmatika


---
Konversi algoritma yang sudah ditulis sebelumnya dengan menggunakan bahasa Pemrograman Python,seperti yang ditunjukkan pada Gambar 2a or Gambar2b.


---

```python
n = int(input("masukkan nilai n = "))
a = int(input("masukkan nilai a = "))
b = int(input("masukkan nilai b = "))
for i in range(n):
  n = i + 1
  un = a + (n - 1) * b
  print("u-",n, "=", un)
sn = n / 2 * (2 * a + (n - 1) * b)
print("sn =",sn)
```
