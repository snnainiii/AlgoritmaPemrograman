##### **1. textbfTuple, akses variabel yang bertipe tuple (jelaskan indeks, dan pengaksesan setiap anggota yang terdapat pada tuple) dan berikan contoh**

**Tuple**

*   tipe data yang digunakan untuk menyimpan lebih dari satu nilai dalam satu variabel sekaligus.
*   memiliki sifat *immutable*, artinya nilai di dalamnya tidak dapat diubah
*   elemant pada tuple dibuka dan ditutup dengan tanda kurung biasa '( )'
*   masing-masing elemen di pisahkan dengan tanda koma ','
*   index dimulai dari 0

**Index**

index adalah cara untuk mendapatkan atau mengetahui informasi mengenai posisi element di dalam tuple

**Mengakses Anggota Tuple**

mengakses tipe data tuple bisa dibilang sama persis dengan mengakses tipe data list

tuple dapat diakses dengan cara *namatuple[index]*

[n] = menampikan suatu index

[a:b] = menampilkan index dari a sampai b-1

[:b] = menampilkan index dari awal sampai b

[-n] = menampilkan index dari element terakhir/index dimulai dari element terakhir

[:-n] = menampilkan index dari a sampai sebelum terakhir







**contoh tuple**

```python
nilai = (60, 70, 80, 90, 100)
print(nilai)
print(nilai, 'memiliki tipe data : ', type(nilai))
```

```python
buah = ('apel', 'anggur', 'pisang', 'semangka')
print(buah)
print(buah, 'memiliki tipe data :', type(buah))
```

```python
mytuple = (60, 'apel', False, 22,7)
print(mytuple)
print(mytuple, 'memiliki tipe data : ', type(mytuple))
```

**contoh pengaksesan tuple**

```python
# contoh [n]
warna = ('hitam', 'putih', 'biru', 'jingga', 'kuning', 'merah', 'hijau')
print(warna[2])
print(warna[4])
print(warna[6])
```

```python
# contoh [a:b]
warna = ('hitam', 'putih', 'biru', 'jingga', 'kuning', 'merah', 'hijau')
print(warna[2:6])
print(warna[0:4])
print(warna[1:3])
```

```python
# contoh [a:]
warna = ('hitam', 'putih', 'biru', 'jingga', 'kuning', 'merah', 'hijau')
print(warna[2:])
print(warna[0:])
print(warna[5:])
```

```python
# contoh [:b]
warna = ('hitam', 'putih', 'biru', 'jingga', 'kuning', 'merah', 'hijau')
print(warna[:4])
print(warna[:1])
print(warna[:6])
```

```python
# contoh [-n]
warna = ('hitam', 'putih', 'biru', 'jingga', 'kuning', 'merah', 'hijau')
print(warna[-2])
print(warna[-5])
print(warna[-3])
```

```python
# contoh [:-n]
warna = ('hitam', 'putih', 'biru', 'jingga', 'kuning', 'merah', 'hijau')
print(warna[:-4])
print(warna[:-1])
print(warna[:-5])
```

#### **2.2 Mendapatkan nama hari dari suatu tanggal**
Buatlah code dengan menggunakan Python mendapatkan nama hari (gunakan data dengan bentuk tuple), dengan ketentuan sebagai berikut

*    user memberikan informasi bahwa tanggal pertama bulan ini, jatuh pada hari tertentu
*    tanggal yang ingin dicari nama harinya, adalah inputan dari user
*    untuk melihat apakah output yang dihasilkan adalah benar, dapat dilihat di calender

Petunjuk : untuk mendapatkan index dari suatu data tuple atau list, bisa digunakan perintah namaT uple.index(anggota).

Contoh output yang dihasilkan dapat dilihat pada Gambar 2.
Screenshot (711).png





```python
# (a) Hari pertama pada bulan tersebut adalah hari Minggu
hari = ('senin','selasa','rabu','kamis','jumat','sabtu','minggu')
info = input('Masukkan informasi, hari pertama bulan ini, jatuh pada hari : ')
tanggal = int(input('masukkan tanggal yang ingin diketahui harinya : '))
for i in range (len(hari)):
    if hari[i] == info:
        a = i - 1
        b = (tanggal + a) % 7
print('Tanggal', tanggal, 'adalah hari' , hari[b])
```

#### **2.3 List Comprehension**
Dengan menggunakan list comprehension, buatlah list ganjil dan list genap dari suatu list yang merupakan inputan dari user, seperti yang ditunjukkan pada contoh Gambar 3

Screenshot (712).png

```python
# (a) input list
list = []
jumlah = int(input('masukan ukuran list = '))
for n in range (jumlah):
    hasil = int(input('masukkan data yang ke-' + str(n) + ': '))
    list.append(hasil)
```

```python
# (b) list ganjil dan genap
ganjil = [i for i in list if i % 2 == 1]
genap = [i for i in list if i % 2 == 0 ]

print('list   = ', list)
print('ganjil = ', ganjil)
print('genap  = ', genap)
```
